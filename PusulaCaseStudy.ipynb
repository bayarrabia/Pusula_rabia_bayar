```python
# Looking at basic statistics of numerical variables
numeric_summary = df.describe()

# Creating a heatmap using seaborn to visualize missing data
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Data Visualization")
plt.show()

numeric_summary
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_columns = ['Kilo', 'Boy']

plt.figure(figsize=(14, 6))

# Histogram chart
for i, col in enumerate(numeric_columns):
    plt.subplot(2, len(numeric_columns), i+1)
    sns.histplot(df[col].dropna(), kde=True, bins=20)
    plt.title(f'{col} Histogram')

# Boxplot chart
for i, col in enumerate(numeric_columns):
    plt.subplot(2, len(numeric_columns), len(numeric_columns) + i+1)
    sns.boxplot(x=df[col].dropna())
    plt.title(f'{col} Boxplot')

plt.tight_layout()
plt.show()

# Examining the frequency distributions of categorical variables
categorical_columns = ['Cinsiyet', 'Ilac_Adi', 'Yan_Etki']
categorical_summary = df[categorical_columns].apply(lambda x: x.value_counts())

# Creating correlation matrix and heatmap with only numerical columns
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Korelasyon Matrisi")
plt.show()

categorical_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the style for plots
sns.set(style="whitegrid")

# Plotting histograms for numerical columns (Kilo and Boy)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Kilo (Weight) distribution
sns.histplot(df['Kilo'].dropna(), bins=20, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Kilo (Weight)')
axes[0].set_xlabel('Weight (kg)')

# Boy (Height) distribution
sns.histplot(df['Boy'].dropna(), bins=20, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Boy (Height)')
axes[1].set_xlabel('Height (cm)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Loading the dataset (assuming you are loading it from an Excel file)
# df = pd.read_excel('side_effect_data.xlsx')  # Insert your actual file path

# Filling in missing values in Gender(Cinsiyet), Province(Il) and Blood Group(Kan Grubu) columns.
categorical_columns = ['Cinsiyet', 'Il', 'Kan Grubu']
imputer_cat = SimpleImputer(strategy='most_frequent')  # Fill with the most frequent value
df[categorical_columns] = imputer_cat.fit_transform(df[categorical_columns])

# Cleaning up the Gender column to make sure 'Female' and 'Male' values are consistent
df['Cinsiyet'] = df['Cinsiyet'].str.strip()  # Remove any extra spaces

# Checking the Gender column after handling missing data
print(df['Cinsiyet'].value_counts())

# Calculate the average weight (Kilo) and height (Boy) for men and women.
# Average for men
mean_kilo_male = df[df['Cinsiyet'] == 'Male']['Kilo'].mean()
mean_boy_male = df[df['Cinsiyet'] == 'Male']['Boy'].mean()

# Average for women
mean_kilo_female = df[df['Cinsiyet'] == 'Female']['Kilo'].mean()
mean_boy_female = df[df['Cinsiyet'] == 'Female']['Boy'].mean()

# Checking if averages are calculated
print(f"Average weight for males: {mean_kilo_male}")
print(f"Average height for males: {mean_boy_male}")
print(f"Average weight for females: {mean_kilo_female}")
print(f"Average height for females: {mean_boy_female}")

# Fill in missing Weight and Height data by gender.
# Filling in missing weight and height for males
df.loc[(df['Cinsiyet'] == 'Male') & (df['Kilo'].isnull()), 'Kilo'] = mean_kilo_male
df.loc[(df['Cinsiyet'] == 'Male') & (df['Boy'].isnull()), 'Boy'] = mean_boy_male

# Filling in missing weight and height for females
df.loc[(df['Cinsiyet'] == 'Female') & (df['Kilo'].isnull()), 'Kilo'] = mean_kilo_female
df.loc[(df['Cinsiyet'] == 'Female') & (df['Boy'].isnull()), 'Boy'] = mean_boy_female

# Checking the results again
print("Number of missing weights (after filling): ", df['Kilo'].isnull().sum())
print("Number of missing heights (after filling): ", df['Boy'].isnull().sum())

# Check the results
print(df[['Cinsiyet', 'Kilo', 'Boy']].head())
```

```python
crosstab = pd.crosstab(df['Cinsiyet'], df['Yan_Etki'])
print(crosstab)

# showing relationship between Cinsiyet and Yan Etki
crosstab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Cinsiyet ve Yan Etki likisi')
plt.show()
```
